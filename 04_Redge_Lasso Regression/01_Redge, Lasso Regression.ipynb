## Model Training

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df =pd.read_csv('Datasets\Algerian_forest_fires_cleaned_dataset.csv')
df.head()
```

```python
df.info()
```

```python
## Drop
df.drop(['day','month','year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
## Encoding
df['Classes']=np.where(df['Classes'].str.contains("not fire"),0,1)
```

```python
df.drop(['Classs'],axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
## Independent And dependent features
X =df.drop('FWI', axis=1)
y =df['FWI']
```

```python
X.head()
```

```python
y.head()
```

```python
## Train Test Split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
## Feature Selection based on correlation
X_train.corr()
```

```python
## Check for multicollinearity
corr = X_train.corr()
sns.heatmap(corr,annot=True)
```

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(X_train,0.85)
```

```python
## drop features when correlation is more than 0.85
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)

X_train.shape, X_test.shape
```

```python
## Feature Scaling OR Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_sclaed =scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
## Box Plot 
sns.boxplot(data=X_train)
plt.title("X_train Before Scaling")
```

```python
sns.boxplot(data=X_train_scaled)
plt.title("X_train After Scaling")
```

Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

linreg = LinearRegression()
linreg.fit(X_train_scaled,y_train)
```

```python
y_pred =linreg.predict(X_test_sclaed)
mae =mean_absolute_error(y_test, y_pred)
score =r2_score(y_test, y_pred)

print("Mean absolute error :", mae)
print("R2 Score :", score)
```

```python
plt.scatter(y_test,y_pred)
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

lasso = Lasso()
```

```python
lasso.fit(X_train_scaled,y_train)
y_pred =lasso.predict(X_test_sclaed)
mae=mean_absolute_error(y_test,y_pred)
score =r2_score(y_test,y_pred)

print("Mean absolute error:",mae)
print("R2 Score :", score)
```

```python
plt.scatter(y_test,y_pred)
```

## Ridge Regression
