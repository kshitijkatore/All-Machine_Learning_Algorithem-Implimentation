```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

%matplotlib inline
```

```python
df = pd.read_csv("Datasets\cardekho_dataset.csv")
```

```python
df.head()
```

#### Data CLeaning

```python
df.isnull().sum()
```

```python
## Remove unwanted columns
df.drop('car_name',axis=1, inplace=True)
df.drop('brand', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['model'].unique()
```

```python
num_features = [feature for feature in df.columns if df[feature].dtype != 'O']
print("Num of Numerical Feature :", len(num_features))

categorical_features =[feature for feature in df.columns if df[feature].dtype =='O']
print("Num of Categorical Feature:", len(categorical_features))
```

```python
## independent And Dependent feature
X = df.drop('selling_price', axis=1)
y =df['selling_price']
```

```python
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X['model']=le.fit_transform(X['model'])  ## LabelEncode Encoding non-numerical value
```

```python
numeric_features =X.select_dtypes(exclude="object").columns
oh_features = ['seller_type', 'fuel_type', 'transmission_type']


from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer

numeric_transformer = StandardScaler()
OneHot_transformer = OneHotEncoder(drop='first')


preprocessor =ColumnTransformer(
    [
        ("StandardScaler",numeric_transformer,numeric_features),
        ("OneHotEncoder",OneHot_transformer,oh_features)
    ],remainder='passthrough'
)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', LinearRegression())
])
```

```python
pipeline.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,r2_score,mean_absolute_error,mean_squared_error

y_pred =pipeline.predict(X_test)

print("R2 Score:",r2_score(y_test,y_pred))
print("Mean Absolute Error :", mean_absolute_error(y_test,y_pred))
print("Mean Squared Error :",mean_squared_error(y_test,y_pred))
```

```python
## Hyperparameter Tunning --> For small datasets Used GridSearchCV
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
params = {
    'model__fit_intercept': [True, False],
    'model__positive': [True, False]  # Only for non-negative regression problems
}
grid_searchCv = GridSearchCV(
    estimator=pipeline,
    param_grid= params,
    scoring='neg_root_mean_squared_log_error',
    cv=5,
    verbose=2,
    n_jobs=1
)

grid_searchCv.fit(X_train,y_train)
```
